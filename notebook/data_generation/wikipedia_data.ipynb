```python
def save_function(page):
    
    text = page.text

    # Initialize PDF structure
    pdf = FPDF()
    pdf.add_page()

    # Path to standard Windows Arial TrueType font
    font_path = os.path.join(os.environ["WINDIR"], "Fonts", "arial.ttf")

    # 1. Register the TrueType font into FPDF (this enables full Unicode support)
    pdf.add_font("ArialUni", fname=font_path, uni=True)

    # 2. Set the font to your newly registered font name
    pdf.set_font("ArialUni", size=12)

    # FPDF2 will now safely render the en-dash, smart quotes, and accents
    pdf.multi_cell(w=0, h=10, txt=text)
    pdf.output(f"./data/wikipedia/{page.title}.pdf")
```

```python
pages_metedata = []

for category in list_of_all_wiki_pages:
    page = wiki.page(category)
    
    # Check if the page actually exists
    if page.exists():
        page_metadata = {
        'title': page.title,
        'fullurl': page.fullurl,
        'summary': page.summary[0:60] 
        }
        pages_metedata.append(page_metadata)
        save_function(page)

    else:
        print("Page not found!")
```
