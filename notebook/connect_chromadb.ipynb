```python
import glob
import os
import sys
home_directory = os.path.expanduser('C:\Program Files\Studying\coding')
package_root = os.path.join(home_directory, 'RAG_project')
sys.path.insert(1,package_root)
os.chdir(sys.path[1])

import chromadb
```

```python
# 1. Initialize the database and tell it where to save data
# This creates a folder named 'my_chromadb' in your current directory
client = chromadb.PersistentClient(path="./my_chromadb")
```

```python
# 2. Create a collection (think of this as a table)
# If it already exists, it will just get the existing one
collection = client.get_or_create_collection(name="my_notes")
```

```python
# 3. Add some text documents
# ChromaDB automatically turns these into vectors using a built-in model!
collection.add(
    documents=[
        "The capital of France is Paris.",
        "Python is a versatile programming language.",
        "A vector database stores data as mathematical coordinates.",
        "The weather in London is often rainy."
    ],
    ids=["id1", "id2", "id3", "id4"]
)
```

```python
# 4. Search the database
results = collection.query(
    query_texts=["What is the capital of France?"], 
    n_results=1 # Give me the top 1 most similar result
)
```
