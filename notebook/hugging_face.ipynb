### first example to inference model

```python
from huggingface_hub import InferenceClient
```

```python
client = InferenceClient()

completion = client.chat.completions.create(
    model="deepseek-ai/DeepSeek-V3-0324",
    messages=[
        {
            "role": "user",
            "content": "answer me few question. Question 1: what model you are ? and when latest information you can provide to me. Question 2: what is ai agent?"
        }
    ],
)
```

```python
print(completion.choices[0].message["content"])
```
